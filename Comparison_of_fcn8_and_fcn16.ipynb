```python
from skimage.color import gray2rgb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from skimage.io import imread, imsave
```

```python
#Loading the Image
image = imread("orginal_1.png")
```

```python
plt.imshow(image)
```

```python
#Loading the annotations
annotation1 = imread("annotation1_fcn16.png")
annotation2 = imread("annotation1_fcn8.png")
```

```python
#Plotting the annotations of FCN-16 and FCN-8
#Note : The left and right ones are obtained from FCN-16 and FCN-8 respectively
plt.subplot(1,2,1)
plt.imshow(annotation1)
plt.subplot(1,2,2)
plt.imshow(annotation2)
```
